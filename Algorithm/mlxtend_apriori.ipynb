```python
import pandas as pd
 
shopping_list = [['豆奶','莴苣'],
        ['莴苣','尿布','葡萄酒','甜菜'],
        ['豆奶','尿布','葡萄酒','橙汁'],
        ['莴苣','豆奶','尿布','葡萄酒'],
        ['莴苣','豆奶','尿布','橙汁']]
 
shopping_df = pd.DataFrame(shopping_list)

shopping_df
```

```python
# shopping_df.stack().groupby(level=0).apply(list).tolist()


def deal(data):
    return data.dropna().tolist()

df_arr = shopping_df.apply(deal, axis=1).tolist()
df_arr
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
df_tf = te.fit_transform(df_arr)
df = pd.DataFrame(df_tf, columns=te.columns_)
```

```python
te.columns_
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values(by='support', ascending=False, inplace=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.9)
association_rules.sort_values(by='lift', ascending=False, inplace=True)
association_rules
```
