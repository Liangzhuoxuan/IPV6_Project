```python
from copy import deepcopy

def _check_orginal(arg):
    counter = 0
    ret = deepcopy(arg)
    if '::' in arg:
        return 0
        splited = arg.split(':')
        l = len([i for i in splited if i != ''])
        if arg.endswith('::'):
            ret = arg.replace('::', (8-l)*':0')
        else:
            ret = arg.replace('::', (8-l)*':0' + ':')
            
    parse_list = ret.split(':')
    cut_position_end = 0 # 记录
    cut_position_start = 0
    for index, setion in enumerate(parse_list):
        try:
            temp = int(setion)
            if 0 <= temp <= 255:
                counter += 1
                if counter == 1:
                    cut_position_start = index
            elif counter >= 4:
                pass
            else:
                counter = 0
        except ValueError:
            if counter >= 4:
                cut_position_end = index - 1
            else:
                counter = 0
    # print(splited[cut_position_start:cut_position_end])
    if counter > 4:
        return 1
    elif counter == 4:
        return 0
    else:
        return -1
```
