```python
def replenish_ip(ip):
    '''
    补全ip中的::为0000
    '''
    org_length = 8
    splited_ = ip.split(':')
    if '' in splited_:
        length = len(splited_) - 1
    else:
        length = len(splited_)
    d = org_length - length
    ip = ip.replace('::', ':0000'*d + ':')
    temp = ip.split(':')[4:]
    print(temp)
    for index, i in enumerate(temp):
        l = len(i)
        if l < 4:
            d = 4 - l
            temp[index] = '0'*d + i

    return ''.join(temp)
```

```python
import numpy as np
import pandas as pd

path = 'D:/address.txt'
with open(path, 'r+', encoding='utf8') as fp:
    content = fp.read().split('\n')
```

```python
content
```

```python
for index, i in enumerate(content):
    temp = list(replenish_ip(i))
    for k,j in enumerate(temp):
        if j == 'a':
            temp[k] = '11'
        elif j == 'b':
            temp[k] = '12'
        elif j == 'c':
            temp[k] = '13'
        elif j == 'd':
            temp[k] = '14'    
        elif j == 'e':
            temp[k] = '15'
        elif j == 'f':
            temp[k] = '16'
    content[index] = temp
content
```

```python
df = pd.DataFrame(content)
df
```

```python
from sklearn.cluster import KMeans

n_clusters = 3
cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(df)
```

```python
y_pred = cluster.labels_
y_pred
```

```python
centroid = cluster.cluster_centers_
centroid
```
